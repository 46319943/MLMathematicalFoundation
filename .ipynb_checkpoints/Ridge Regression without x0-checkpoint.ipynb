```python
import numpy as np

X_train = np.loadtxt('https://www.csie.ntu.edu.tw/~htlin/mooc/datasets/mlfound_algo/hw4_train.dat')
Y_train = X_train[:, [-1]]
X_train = X_train[:, :2]
X_test = np.loadtxt('https://www.csie.ntu.edu.tw/~htlin/mooc/datasets/mlfound_algo/hw4_test.dat')
Y_test = X_test[:, [-1]]
X_test = X_test[:, :2]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_train[:, 0], Y_train)
```

```python
Lambda = 10
w = np.matmul(
        np.linalg.inv(
            np.matmul(X_train.T, X_train) + Lambda * np.identity(2)
        ),
        np.matmul(X_train.T, Y_train)
    )
```

```python
y = np.matmul(X_train, w)
```

```python
plt.scatter(X_train[:, 0], y)
```
